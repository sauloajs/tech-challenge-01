```python
df= pd.read_csv('/content/ExpVinho (1).csv', sep =';')
df = df.set_index('País')
df
```

```python
df_volume = df.iloc[:, ::2]
df_volume_15_anos =df_volume.iloc[:,-15:]
df_volume_15_anos
```

```python
# Calcular o volume total por país para cada ano
total_volume_by_country = df_volume.sum(axis=0)

# Ordenar os países pelo volume total decrescente
sorted_countries = total_volume_by_country.sort_values(ascending=False)

# Calcular a porcentagem cumulativa por país
cumulative_percentage = sorted_countries.cumsum() / total_volume_by_country.sum()

# Filtrar os países que representam 80% do volume total
selected_countries = cumulative_percentage[cumulative_percentage <= 0.8].index

# Filtrar as colunas do DataFrame original apenas para os países selecionados
df_selected_countries = df_volume[selected_countries]

# Plotar o gráfico de barras empilhadas
plt.figure(figsize=(20, 6))

bottom = np.zeros(len(df_selected_countries.columns))

for year in df_selected_countries.index:
    plt.bar(selected_countries, df_selected_countries.loc[year], bottom=bottom, label=year)
    bottom += df_selected_countries.loc[year]

plt.xlabel('Country')
plt.ylabel('Volume')
plt.title('Stacked Volume of Countries Representing 80%')
plt.legend()
plt.show()
```

```python
# Calcular o volume total por país para cada ano
total_volume_by_year = df_volume.sum(axis=0)

# Calcular a porcentagem cumulativa do volume total por ano
cumulative_percentage = total_volume_by_year.cumsum() / total_volume_by_year.sum()

# Filtrar os anos que representam 80% do volume total
selected_years = cumulative_percentage[cumulative_percentage <= 0.8].index

# Filtrar as colunas do DataFrame original apenas para os anos selecionados
df_selected_years = df_volume[df_volume.columns.intersection(selected_years)]

# Plotar o gráfico de barras empilhadas
plt.figure(figsize=(12, 6))

bottom = np.zeros(len(df_selected_years.columns))

for country in df_selected_years.index:
    bars = plt.bar(df_selected_years.columns, df_selected_years.loc[country], bottom=bottom, label=country)
    bottom += df_selected_years.loc[country]

plt.xlabel('Ano')
plt.ylabel('Volume')
plt.title('Stacked Volume of Countries Representing 80%')

plt.legend()

plt.show()
```

```python
total_volume = df_volume_15_anos.sum().sum()

# Calcule a soma de cada país e ordene do maior para o menor
country_sums = df_volume_15_anos.sum().sort_values(ascending=False)

# Calcule a soma cumulativa
cumulative_sums = country_sums.cumsum()

# Selecione os países que correspondem a 80% do total
selected_countries = cumulative_sums[cumulative_sums <= 0.8 * total_volume].index

# Selecione apenas esses países do DataFrame
df_selected = df_volume_15_anos[selected_countries]

# Agora, crie o gráfico
df_selected.plot(kind='bar', stacked=True, figsize=(10, 7))

plt.title('Volume de Vinho Exportado nos Últimos N Anos (80% do Volume Total)')
plt.xlabel('Ano')
plt.ylabel('Volume de Vinho Exportado')
plt.legend(title='País')

plt.show()
```

```python
# Crie um DataFrame vazio para armazenar os dados selecionados
df_selected = pd.DataFrame()

# Para cada coluna no DataFrame
for year in df_volume_15_anos.columns:
    # Calcule a soma total para o ano
    total_volume = df_volume_15_anos[year].sum()

    # Ordene os países pelo volume de exportação e calcule a soma cumulativa
    sorted_volumes = df_volume_15_anos[year].sort_values(ascending=False)
    cumulative_sums = sorted_volumes.cumsum()

    # Selecione os países que correspondem a 80% do total
    selected_countries = cumulative_sums[cumulative_sums <= 0.8 * total_volume].index

    # Adicione esses países e os respectivos volumes ao DataFrame df_selected
    df_selected = pd.concat([df_selected, df_volume_15_anos.loc[selected_countries, year]], axis=1)

# Renomeie as colunas do DataFrame df_selected com os respectivos anos
df_selected.columns = df_volume_15_anos.columns

# Crie o gráfico de barras empilhadas
df_selected.T.plot(kind='bar', stacked=True, figsize=(10, 7))

plt.title('Volume de Vinho Exportado nos Últimos 15 Anos (80% do Volume Total por Ano)')
plt.xlabel('Ano')
plt.ylabel('Volume de Vinho Exportado')
plt.legend(title='País', loc='upper left', bbox_to_anchor=(1.0, 1.0))  # Move the legend outside of the plot

plt.show()
```

```python

# Define the thresholds for each axis
x_threshold = df_crossdata17['GDP per capita, PPP (constant 2017 international $)'].median()
y_threshold = df_crossdata17['Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'].median()

# Retrieve the data for the scatter plot
consumo_alcool = df_crossdata17['Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)']
gdp_per_capita = df_crossdata17['GDP per capita, PPP (constant 2017 international $)']
populacao = df_crossdata17['Population (historical estimates)']

# Create the scatter plot with quadrants
plt.figure(figsize=(10, 6))

# High-High quadrant (above x_threshold and above y_threshold)
plt.scatter(gdp_per_capita[consumo_alcool > y_threshold], consumo_alcool[consumo_alcool > y_threshold], s=populacao[consumo_alcool > y_threshold]/100000, alpha=0.7, label='High-High')

# High-Low quadrant (above x_threshold and below y_threshold)
plt.scatter(gdp_per_capita[consumo_alcool <= y_threshold], consumo_alcool[consumo_alcool <= y_threshold], s=populacao[consumo_alcool <= y_threshold]/100000, alpha=0.7, label='High-Low')

# Low-High quadrant (below x_threshold and above y_threshold)
plt.scatter(gdp_per_capita[gdp_per_capita <= x_threshold], consumo_alcool[gdp_per_capita <= x_threshold], s=populacao[gdp_per_capita <= x_threshold]/100000, alpha=0.7, label='Low-High')

# Low-Low quadrant (below x_threshold and below y_threshold)
plt.scatter(gdp_per_capita[(consumo_alcool <= y_threshold) & (gdp_per_capita <= x_threshold)], consumo_alcool[(consumo_alcool <= y_threshold) & (gdp_per_capita <= x_threshold)], s=populacao[(consumo_alcool <= y_threshold) & (gdp_per_capita <= x_threshold)]/100000, alpha=0.7, label='Low-Low')

# Set labels and title for the plot
plt.xlabel('GDP per capita (PPP)')
plt.ylabel('Total alcohol consumption per capita')
plt.title('Consumo de álcool vs. GDP per capita')

# Add a legend
plt.legend()

plt.grid(True)
plt.show()
```

```python
df= pd.read_csv('/content/ExpVinho (en).csv', sep =';')
df = df.set_index('País')
df
```

```python
df_valor = df.iloc[:, 1::2]
df_valor_15_anos =df_valor.iloc[:,-15:]
df_valor_15_anos
```

```python
df_trade = pd.read_excel('/content/TRADEMAP_20ANOS.xlsx', skiprows=13)
df_trade = df_trade.set_index('Importers')
df_trade
```

```python
paises_selecionados = list(set(df_valor_15_anos.index).intersection(df_trade.index))
paises_selecionados
```

```python
for pais in paises_selecionados:
    df_pais_valor = df_valor_15_anos.loc[pais]
    df_pais_trade = df_trade.loc[pais]
    
    exportacoes_brasileiras = df_valor_15_anos.loc[pais].sum()
    importacoes_totais = df_pais_trade.sum()
    
    market_share = (exportacoes_brasileiras / importacoes_totais) * 100
    
    print(f"Market share das exportações brasileiras para {pais}: {market_share}%")
```

```python
market_share_data = []

# Iterate over the selected countries
for pais in paises_selecionados:
    market_share_pais = []
    
    # Iterate over the years
    for ano in df_valor_15_anos.columns:
        ano_int = int(ano)
        exportacoes_pais = df_valor_15_anos.loc[pais, ano]
        importacoes_pais = df_trade.loc[f'{pais}', ano_int]
        
        # Check if the import value is NaN
        if np.isnan(importacoes_pais):
            market_share = np.nan
        else:
            market_share = (exportacoes_pais / importacoes_pais) * 100
        
        market_share_pais.append(market_share)
    
    market_share_data.append(market_share_pais)

# Create a DataFrame with the market share data
df_market_share = pd.DataFrame(market_share_data, columns=df_valor_15_anos.columns)

# Add the country column to the DataFrame
df_market_share.insert(0, 'País', paises_selecionados)

# Display the resulting DataFrame
df_market_share.head(10)
```
