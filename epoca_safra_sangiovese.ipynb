```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FULL_PATH = 'C:\\Users\\caval\\Documents\\tech-challenge-01'
```

```python
dataset = pd.read_csv(f'{FULL_PATH}\\content\\dados-clima.csv')
dataset
```

```python
data = dataset.drop(columns=['Unnamed: 7', 'Codigo Estacao', 'Data Inicial', 'Data Final', 'Situacao', 'Periodicidade da Medicao', 'Latitude', 'Longitude', 'Altitude'], axis=1)
data.head()
```

```python
data['Nome'] = data['Nome'].str.strip()
```

```python
data['Nome'].unique()
```

```python
bento_goncalves = data.loc[data['Nome'] == 'BENTO GONCALVES']
```

```python
bento_goncalves = bento_goncalves.iloc[:-17, :]
```

```python
bento_goncalves.iloc[-12:, :]
```

```python
sns.barplot(data=bento_goncalves.iloc[-12:, :])
```

```python
bento_goncalves.columns = ['mes_ano', 'precipitacao_dias', 'precipitacao_total', 'pressao_media_mensal', 'temperatura_media_mensal', 'velocidade_vento_mensal_max', 'velocidade_vento_mensal', 'cidade']
```

```python
bento_goncalves.set_index('mes_ano', inplace=True)
```

```python
bento_goncalves.reset_index(inplace=True)
```

```python
# bento_goncalves['precipitacao_dias'] = bento_goncalves['precipitacao_dias'].astype(int)
bento_goncalves['temperatura_media_mensal'] = bento_goncalves['temperatura_media_mensal'].str.replace(',', '.').astype(float)
bento_goncalves['pressao_media_mensal'] = bento_goncalves['pressao_media_mensal'].str.replace(',', '.').astype(float)
bento_goncalves['precipitacao_total'] = bento_goncalves['precipitacao_total'].str.replace(',', '.').astype(float)
```

```python
plt.figure(figsize=(12, 6))

ax = sns.barplot(data=bento_goncalves.iloc[-12:, :], x='mes_ano', y='precipitacao_dias')
ax.tick_params(axis='x', rotation=45)

ax.set_title('Precipitação de dias por mês')
```

```python
plt.figure(figsize=(12, 6))

ax = sns.barplot(data=bento_goncalves.iloc[96:108, :], x='mes_ano', y='precipitacao_dias')
ax.tick_params(axis='x', rotation=45)

ax.set_title('Precipitação de dias por mês')
```

```python
bg_2016 = bento_goncalves.iloc[96:108, :]
bg_2021 = bento_goncalves.iloc[-12:, :]
```

```python
plt.figure(figsize=(12, 6))

ax = sns.lineplot(data=bg_2016.iloc[:5, :], x='mes_ano', y='precipitacao_dias')
ax.tick_params(axis='x', rotation=45)
ax.set_title('Precipitação de dias por mês')
ax2 = ax.twinx()
ax = sns.lineplot(data=bg_2021.iloc[:5, :], x='mes_ano', y='precipitacao_dias', ax=ax2, marker='o', color='green')
```
