```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
columns = []
columns.extend(['id', 'category', 'sub_category'])
columns.extend(range(1970, 2022))
columns
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/sauloajs/tech-challenge-01/main/content/Comercio.csv", sep=';', index_col=False, names=columns)
dataset.head(20)
```

```python
dataset.columns
```

```python
def getWineCategories(columns):
  categories = []
  
  for column in columns:
    if (column.isupper()):
      categories.append(column)

  return categories
```

```python
columns = dataset['category']
wine_categories = getWineCategories(columns)
wine_categories
```

```python
total_sales_by_categories = dataset.set_index('category')
total_sales_by_categories = total_sales_by_categories.T[wine_categories]
```

```python
total_sales_by_categories.drop('id', 0, inplace=True)
```

```python
total_sales_by_categories
```

```python
total_sales_by_categories.T
```

```python
from datetime import date

def to_date(column):
  year = int(column)
  return date(year, 1, 1)
```

```python
total_sales_by_categories = total_sales_by_categories.drop('sub_category', 0)
```

```python
total_sales_by_categories.index
```

```python
total_sales_by_categories = total_sales_by_categories.T.reset_index()
total_sales_by_categories
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(data=total_sales_by_categories, x='category', y=2019)
ax.tick_params(axis='x', rotation=55)
plt.grid(visible=True, color='gray', linestyle='--', lw=0.2)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
```

```python
table_wine = total_sales_by_categories.set_index('category').iloc[:, -30:]
axis = table_wine.T.plot(figsize=(16,8))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.xticks(rotation=28)
plt.grid(visible=True, color='gray', linestyle='--', lw=0.2)
plt.title("Valores Comercializados nos últimos 30 anos por categoria majoritária")
plt.ylim(0, round(max(table_wine.max()) * 1.075))

leg = plt.legend(loc='upper right')

plt.draw()

bb = leg.get_bbox_to_anchor().transformed(axis.transAxes.inverted())

xOffset = 0.275
bb.x0 += xOffset
bb.x1 += xOffset
leg.set_bbox_to_anchor(bb, transform = axis.transAxes)

plt.show()
```

```python
interested_categories = ['vm_Tinto', 'vm_Rosado', 'vm_Branco']
table_wines_by_categories = dataset.query("category in @interested_categories")
table_wines_by_categories
```

```python
fine_quality_table_wines = fine_quality_table_wines.drop(['id', 'sub_category'], 1)
fine_quality_table_wines
```

```python
fine_quality_table_wines = fine_quality_table_wines.set_index('category')
fine_quality_table_wines
```
