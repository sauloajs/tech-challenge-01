```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 125)
pd.set_option('display.max_columns', 52)
```

```python
FULL_PATH = "C:\\Users\\caval\\Documents\\tech-challenge-01"
```

```python
dataset = pd.read_csv(f"{FULL_PATH}\\content\\ExpVinho.csv", sep=';')

dataset.head(5)
```

```python
year_columns = dataset.columns[2:].str[:4]
base_columns = dataset.columns[:2]
```

```python
list_columns = []
list_columns.extend(base_columns)
list_columns.extend(year_columns)
list_columns
```

```python
dataset.columns = list_columns
dataset.head()
```

```python
dataset = dataset.groupby(level=0, axis=1).sum()
dataset
```

```python
dataset.drop('Id', axis=1, inplace=True)
dataset
```

```python
dataset.set_index('País', inplace=True)
dataset.head()
```

```python
sns.lineplot(data=dataset.T)
```

```python
dataset.loc[:, 'total_exported'] = dataset.sum(numeric_only=True, axis=1)
```

```python
dataset.sort_values("total_exported", ascending=False, inplace=True)
dataset.drop("total_exported", axis=1, inplace=True)
```

```python
dataset.head(10)
```

```python
plt.figure(figsize=(12,8))
axis = sns.lineplot(data=dataset.iloc[:, :30].head(5).T, dashes=False)
axis.tick_params(axis='x', rotation=55)
```
