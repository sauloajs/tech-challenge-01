```python
def get_grape_categories(columns):
    categories = []
    
    for column in columns:
        if (column.isupper()):
            categories.append(column)
            
    return categories
```

```python
def get_interested_grapes(wine_grapes):
    grapes = []
    for grape in wine_grapes:
        blend = grape.find("-")
        if (blend != -1):
            grapes.append(grape[:blend].strip())
            grapes.append(grape[blend+1:].strip())
        else:
            grapes.append(grape.strip())
    return grapes
```

```python
interested_grapes = set(get_interested_grapes(valid_scores['grape']))
interested_grapes
```

```python
utilized_red_grapes = red_grapes.query('cultivar in @interested_grapes')
utilized_red_grapes.isnull().sum()
```

```python
utilized_grapes
```

```python
fig = go.Figure()

for step in utilized_red_grapes:
    fig.add_trace(
        go.Scatter(
            visible=False,
            line=dict(color="#00CED1", width=6),
            name=str(step),
            y=utilized_red_grapes[step],
            x=utilized_red_grapes.index
        )
    )

steps = []
for i in range(len(fig.data)):
    step = dict(
        method="update",
        args=[{"visible": [False] * len(fig.data)},
              {"title": "Slider switched to step: " + str(i)}], 
    )
    step["args"][0]["visible"][i] = True
    steps.append(step)

sliders = [dict(
    active=10,
    currentvalue={"prefix": "Ano: "},
    pad={"t": 30},
    steps=steps
)]

fig.update_layout(
    sliders=sliders
)

fig.show()
```
